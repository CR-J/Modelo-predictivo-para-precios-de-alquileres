Cargo las librer√≠as:

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Cargo los datasets:

```python
X_train = pd.read_csv('X_train_ns.csv')
del X_train['Unnamed: 0']
y_train = pd.read_csv('log_y_train.csv')
del y_train['Unnamed: 0']
y_test = pd.read_csv('log_y_test.csv')
del y_test['Unnamed: 0']
```

## Histogramas de la variable objetivo, log("Price")

El primer histograma muestra como la variable objetivo se acumula en ciertas columnas. Esto es porque los precios tienden a ser enteros, y tienden a redondearse hacia ciertos valores, por ejemplo 99 en lugar de 100:

```python
objetivo = list(list(np.array(y_train).reshape(1,-1))[0])
sns.histplot(objetivo,binwidth=0.01)
plt.xlabel('Logaritmo del precio por noche')
plt.show()
```

Repito el histograma con barras algo m√°s anchas para que el histograma no se vea afectado por estas acumulaciones:

```python
sns.histplot(objetivo,binwidth=0.25)
plt.xlabel('Logaritmo del precio por noche en train')
plt.show()
```

Compruebo que el logaritmo del precio se reparte de forma similar en train y en test comparando el histograma anterior (sobre train) con el siguiente (sobre test):

```python
objetivot = list(list(np.array(y_test).reshape(1,-1))[0])
sns.histplot(objetivot,binwidth=0.25)
plt.xlabel('Logaritmo del precio por noche en test')
plt.show()
```

## Relaci√≥n de la variable objetivo frente al resto de variables

Los scatterplots de la variable objetivo log("Price") frente a las distintas variables predictoras me aportan informaci√≥n sobre sus relaciones.

Adem√°s, he a√±adido las correlaciones entre log("Price") y las variables predictoras.

Los gr√°ficos est√°n ordenados desde las variables predictoras con mayor informaci√≥n mutua a las variables predictoras con menor informaci√≥n mutua, con lo que se puede ver que la informaci√≥n mutua y la correlaci√≥n no dan la misma importancia a las distintas variables: la correlaci√≥n mide relaciones lineales, mientras que la informaci√≥n mutua mide cualquier tipo de relaci√≥n, lineal o no.

```python
# El siguiente c√≥digo me permitir√° emplear negrita al usar print:

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
    
for var in X_train.columns:
    print("La correlaci√≥n es de",
          color.BOLD+str(round(np.corrcoef(objetivo,list(np.array(X_train[var])))[0,1],2))+color.END,
         "entre la variable",color.BOLD+str(var)+color.END,"y el logaritmo del precio por noche.")
    sns.scatterplot(x=objetivo,
                y=list(np.array(X_train[var])),alpha=0.5)
    plt.xlabel('Logaritmo del precio por noche')
    plt.ylabel(var)
    plt.show()
```
